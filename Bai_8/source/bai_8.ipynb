```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
from run_model import run
```

```python
train_data = pd.read_csv(r'C:\Th-th-ch-BAI-Labs\Bai_8\data\UNSW_NB15_training-set.csv')
test_data = pd.read_csv(r'C:\Th-th-ch-BAI-Labs\Bai_8\data\UNSW_NB15_testing-set.csv')
```

```python
rf = RandomForestClassifier(random_state = 42)
gbc = GradientBoostingClassifier(random_state = 42)
abc = AdaBoostClassifier(random_state = 42)
lr = LogisticRegression(random_state = 42)
knn = KNeighborsClassifier(n_neighbors = 2)
gnb = GaussianNB()
dt = DecisionTreeClassifier(random_state = 42)
xbg = XGBClassifier()
svm = SVC(random_state = 42)
lgb = LGBMClassifier(random_state = 42)
```

```python
run([rf, gbc, abc, lr, knn, gnb, dt, xbg, svm, lgb], train_data, test_data)
```
